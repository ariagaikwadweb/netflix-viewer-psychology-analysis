```python
import matplotlib.pyplot as plt

top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))

top_countries.plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content", fontsize=16)

plt.xlabel("Country")

plt.ylabel("Number of Titles")

plt.xticks(rotation=45)

plt.show()
```

## Insight

The United States dominates Netflix content production, while countries like India, the United Kingdom, and South Korea also contribute significantly to the platform’s global catalogue.

This reflects Netflix’s strategy of balancing Hollywood dominance with growing international content demand.

```python
import matplotlib.pyplot as plt

# Count content types
content_counts = df['type'].value_counts()

# Create figure
plt.figure(figsize=(8,5))

# Bar chart
plt.bar(content_counts.index, content_counts.values)

# Titles and labels
plt.title("Netflix Content Distribution", fontsize=16)

plt.xlabel("Content Type")

plt.ylabel("Number of Titles")

# Show values on bars
for i, value in enumerate(content_counts.values):
    plt.text(i, value + 100, str(value), ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')

plt.title("Movies vs TV Shows on Netflix")

plt.xlabel("Content Type")

plt.ylabel("Count")

plt.show()
```

Netflix focuses more heavily on movies than TV shows, suggesting that movie content still dominates the platform's overall catalogue.
